### Importing all the necessary libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
```

### Reading the .csv file

```python
sample=pd.read_csv('SampleSuperstore.csv')
sample.head(10)
```

```python
sample.shape
```

```python
sample.info()
```

Checking for null values

### Checking No. of Null values in the dataset

```python
sample.isnull().sum()
```

```python
sample.describe()
```

### Checking No. of Duplicate values in the dataset

```python
sample.duplicated().sum()
```

### Deleting all the Duplicate Values

```python
sample.drop_duplicates()
```

### Plot depicting the Shipment Count across various Ship Modes

```python
plt.figure(figsize=(9,5))
plt.style.use('bmh')
sns.countplot(x='Ship Mode',data=sample)
plt.ylabel('No. of Shipments')
```

#### Standard Class was the most preferred Shipment Mode and Same Day was the least preferred option.

# 

### Dropping all the Unnecessary Columns

```python
sample.drop(['Postal Code'],axis=1,inplace=True)
```

```python

sample.drop([ 'Ship Mode'],axis=1,inplace=True)
sample.columns
```

### Plotting Pairwise relationship b/w columns

```python
sns.pairplot(data=sample, hue='Category')
```

### HeatMap 

```python
samp_heat=sample.corr()
sns.heatmap(samp_heat, annot=True, cmap='Blues')
```

#### Least Correlated: Discount and Quantity
#### Most Correlated : Profit and Sales

# 

### Total Sales and Profit 

```python
print("The TOTAL SALES in the dataset is {}".format(sample["Sales"].sum()))
print("The TOTAL PROFIT in the dataset is {}".format(sample["Profit"].sum()))
```


### Graph showing Sales across different Categories

```python
plt.figure(figsize=(9,5))
plt.bar(sample['Category'],sample['Sales'])
plt.xlabel('Cateories')
plt.ylabel('Sales')
plt.show()
```

### Count of Segment by Category and Segment

```python
plt.figure(figsize=(9,5))
sns.countplot(x='Category',hue='Segment',data=sample)
plt.ylabel('Count of Segment')
plt.title('Count of Segment by Category and Segment')
```



### JointPlot depicting Sales across Regions

```python
sns.jointplot(x='Region', y='Sales', data=sample)
```








### CountPlot depicting the No. of Sub-Categories

```python
plt.figure(figsize=(16,9))
sns.countplot(sample['Sub-Category'], palette='dark')

plt.title('Count of Sub-Category', fontsize=15)
```

# 

### LinePlot depicting relation b/w Discount and Profit 

```python

sns.lineplot(y="Profit", x="Discount",data=sample,color='red')
```

# 

### Analysing the Total Sales, Profit, Discount of all the Sub-Categories

```python
sample_subcategory=sample.groupby(['Sub-Category'])[['Sales', 'Profit', 'Discount']].sum()
sample_subcategory.head(18)
```

#### Phones had the highest sales among all Sub-Categories followed by Chairs, Storage, Tables, etc. Copiers made the highest profit. Bookcases made the least profit but made good Sales. The cost of Bookcases can be increased to increase the profit.

# 

### PieChart depicting distribiution of Segment

```python
seg=sample["Segment"].value_counts()
seg
```

```python
seg.plot(kind="pie", figsize=(5,5),autopct='%1.1f%%',fontsize=13, shadow=True)
```

#### Consumer segment has highest share followed by Corporate and Home Office. 

# 

### Segment-wise Profit & Sales

```python
plt.style.use('seaborn-dark')
seg_salesNprofit=sample.groupby("Segment")[["Sales", "Profit"]].sum()
seg_salesNprofit.plot(kind="bar", figsize=(8,5))
plt.ylabel("Sales & Profit")
plt.title("Segment: Profit & Sales", fontsize=15)

```

# 

### State-wise value count

```python
state_count=sample["State"].value_counts()
state_count
```

### State-wise Distribution of Consumers

```python
plt.style.use('seaborn-bright')
state_count.plot(kind="bar", figsize=(15,8))
plt.xlabel("States")
plt.ylabel("No. of Consumers")
plt.title("Distribution of Consumers in each State")
```

# 

### Summing up the Sales across various Categories

```python
category_sales_sum=sample.groupby("Category").Sales.sum()
category_sales_sum
```

```python
label=['Furniture','Office Supplies','Technology']
plt.style.use('fivethirtyeight')
category_sales_sum.plot(kind="pie", figsize=(7,7),labels=label,autopct='%1.1f%%',fontsize=13, shadow=True)
```

### Pie-Chart depicting the proportion of Sales across the 3 Categories 

#### Technology has the highest Sales followed by Furniture and Office Supplies. Also, all the Categories contribute nearly the same amount towards total sales.


# 

### Profit & Sales across various Categories 

```python
sample.groupby('Category')['Sales','Profit'].agg(sum).plot(kind='bar')
plt.style.use('seaborn-colorblind')
plt.ylabel('Profit & Sales')
plt.show()
```

### Furniture is making the least profit among the 3 categories

# 

### Summing up the Profit across different Regions

```python
region_profit_sum=sample.groupby("Region").Profit.sum()
region_profit_sum
```

```python
plt.style.use('bmh')
standout=[0,0,0,0.1]
plt2=region_profit_sum.plot(kind="pie",explode=standout, figsize=(7,7),autopct='%1.1f%%',fontsize=12, shadow=True)
```

#### The West region made the highest Profit, followed by East, South and Central Region.
